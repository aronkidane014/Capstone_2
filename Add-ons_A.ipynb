```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load cleaned datasets
customer_list = pd.read_csv('cleaned_customer_list.csv')
transaction_data = pd.read_csv('time_filtered_transaction_data.csv')

# Merge datasets on 'CustID' (or the common column for customer identification)
combined_data = pd.merge(transaction_data, customer_list, left_on='CustID', right_on='cust_id')

# Preview the combined data
print("Combined Customer and Transaction Data:")
print(combined_data.head())
```

```python

customer_spending = combined_data.groupby('CustID')['OrderTotal'].sum().sort_values(ascending=False)
print("Total Spending by Customer:")
print(customer_spending.head(10))
```

```python
frequent_purchases = combined_data.groupby(['CustID', 'ProductName']).size().sort_values(ascending=False)
print("Most Frequent Purchases by Customer:")
print(frequent_purchases.head(10))
```

```python
# Define customer segments based on total spending
def segment_customer(spending):
    if spending > 10000:
        return 'High Value'
    elif spending > 5000:
        return 'Medium Value'
    else:
        return 'Low Value'

customer_segments = customer_spending.apply(segment_customer)
combined_data = combined_data.merge(customer_segments.rename('Customer Segment'), left_on='CustID', right_index=True)
print("Customer Segments:")
print(combined_data[['CustID', 'Customer Segment']].drop_duplicates().head(10))
```

```python
# Plot total spending by customer
customer_spending.head(10).plot(kind='bar', title='Top 10 Customers by Spending', xlabel='Customer ID', ylabel='Total Spending')
plt.show()
```

```python
# Plot customer segment distribution
combined_data['Customer Segment'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Customer Segments')
plt.show()
```

### Customer Transaction Analysis

To extend the project, we combined the customer list and transaction data to gain further insights into customer transactions and behaviors.

#### Key Analyses

1. **Total Spending by Customer**: This analysis identifies high-value customers based on their cumulative spending.
2. **Average Order Value by Customer**: This metric provides insights into typical spending per transaction for each customer.
3. **Frequent Purchases**: Shows the most commonly purchased products by each customer, highlighting specific preferences.
4. **Customer Segments**: Customers are segmented based on their spending into high, medium, and low-value groups, allowing for targeted marketing strategies.

#### Visualizations

1. **Top 10 Customers by Spending**: A bar chart showcasing the highest-spending customers.
2. **Customer Segment Distribution**: A pie chart visualizing the distribution of customers in each spending segment.

This additional analysis deepens our understanding of customer behaviors, enabling more effective marketing and customer retention strategies.
